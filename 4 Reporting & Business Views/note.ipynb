

# **Layer 4 – Reporting & Business Views**

**Purpose:**

* Provide **ready-to-use analytics** for business users and dashboards.
* Use **clean, friendly names** and pre-aggregated metrics.
* No raw data changes — read-only layer.

**Steps / Objects:**

1. **Reporting Views**

   * Examples:

     * `vw_daily_sales` → daily revenue trends
     * `vw_top_products` → top-selling products
     * `vw_customer_segments` → customer segmentation / value
   * Purpose: answer **common business questions** quickly.
   * Built on **fact_sales** and **dimension tables**.

2. **Monitoring / Audit Table**

   * Table: `production.load_audit`
   * Tracks:

     * Table / layer name
     * Load timestamp
     * Row counts
     * Success / failure indicator
   * Purpose: ensures **pipeline health** and traceability

**Key Notes:**

* This layer does **not store new data**, only aggregates and views.
* Speeds up queries for dashboards / BI tools.
* Ensures **business-friendly names** and simplified logic.
* Supports **daily monitoring** of data loads and quality.
